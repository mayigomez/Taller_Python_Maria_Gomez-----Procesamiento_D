<h3>Introducción </h3>


 <div style="text-align: justify;">Las clases permiten definir estructuras que agrupan datos y funciones relacionadas en una sola entidad. En Python, una clase sirve como plantilla para crear objetos que contienen atributos y métodos.
En este cuaderno se introducen los conceptos básicos de la programación orientada a objetos, incluyendo la creación de clases, la inicialización de atributos y el uso de métodos.</div>

Dado que en el cuaderno propuesto por el profesor no se encontro ningún cuestionario del tema hice unos cuántos ejemplos

<h4> Proposición y Solución de ejemplo</h4>

```python
# Definición de la clase
class Estudiante:
    def __init__(self, nombre, semestre, carrera):
        # Atributos de la clase
        self.nombre = nombre
        self.semestre = semestre
        self.carrera = carrera

    # Método de la clase
    def mostrar_informacion(self):
        print("Nombre:", self.nombre)
        print("Semestre:", self.semestre)
        print("Carrera:", self.carrera)

    # Otro método
    def avanzar_semestre(self):
        self.semestre += 1
        
        print("El estudiante ha avanzado al semestre", self.semestre)


# Creación de un objeto a partir de la clase
estudiante1 = Estudiante("María", 5, "Ciencia de Datos")

# Uso de los métodos del objeto
estudiante1.mostrar_informacion()
estudiante1.avanzar_semestre()
```
