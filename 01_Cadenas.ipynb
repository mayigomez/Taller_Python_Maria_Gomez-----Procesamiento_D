<h4> Solución del Examen sobre Cadenas </h4>

¿Cual es el valor de la variable <code>A</code> después de ejecutar el siguiente código?

```python
A = "1" #Se asigna el elemento "1" a la variable A
```

El nuevo valor de la variable A tiene que ser <code>1</code> ya que la cadena, tiene como contenido el dígito 1, y es el mismo que se le asigna a la variable A; por lo que al intentar imprimir el contenido de A se visualiza...

```python
print(A) #Se imprime la variable A, lo que da como resultado "1"
```

... el dígito <code>1</code>, ya que fue el que se asignó anteriormente a la variable <code>A</code>.

¿Cual es el valor de la variable <code>B</code> después de ejecutar el siguiente código?

```python
B = "2" #Se asigna el elemento "2" a la variable B
```

Así como en el ejercicio anterior, el dígito "2" que se encuentra en la cadena es el contenido que se le asigna a la variable B; por cual, al volver a imprimir se verifica que el valor guardado en la variable <code>B</code> es 2.

```python
print(B) #Se imprime la variable B,
```

¿Cual es el valor de la variable <code>C</code> después de ejecutar el siguiente código?

```python
C = A + B #Se crea una nueva variable C, cuyo contenido es la concatenación de las variables A y B
```

El valor de la variable C va a ser la concatenación de las dos cadenas anteriores, A y B, que a través de la suma formaron una nueva cadena a partir de la combinación de ambas.

```python
print(C) #Se imprime la concatenación de las variables A y B, lo que da como resultado "12"
```

Y cuando se imprime, se logra percibir que el contenido guardado es el equivalente a la cadena A es decir, 1; y a la cadena B es decir, 2.

En la variable <code>D</code> haz uso del slicing para imprimir solo los primeros tres elementos:

```python
D = "ABCDEFG" #Se guardan todos lo elementos de la cadena en la variable 'D'
print(D[0:3]) #Se imprime solo los 3 primeros elementos de la cadena 'D'
```

Usa un valor de stride de 2 para imprimir cada segundo elemento de la cadena <code>E</code>:

```python
E = 'clocrkr1e1c1t' #Se guardan todos lo elementos de la cadena en la variable 'E'
print(E[::2]) #Se imprime cada segundo elemento de la cadena 'E'
```

Imprime una barra invertida:

```python
S ='\\' #Se guarda la barra invertida en la variable 'S'
print(S) #Se imprime la variable 'S'
```

Convierte el mayúsculas el contenido de la variable <code>F</code>:

```python
F = "You are wrong" #Se guarda la cadena "You are wrong" en la variable 'F'
print("Antes de pasar todas las letras a mayúsculas de la palabra: ", F) #Se imprime la cadena 'F' en su estado inicial
M = F.upper() #Se guarda la variable 'F' modificada a mayúsculas en la variable 'M'
print("Después de pasar todas las letras a mayúsculas de la palabra: ", M) #Se imprime la variable 'M' que tiene la nueva cadena 'F' modificada a mayúsculas
```

En la variable <code>G</code>, encuentra el primer índice de la sub-cadena <code>snow</code>:

```python
# Escribe abajo tu código y presiona Shift+Enter para ejecutarlo

G = "Mary had a little lamb Little lamb, little lamb Mary had a little lamb \
Its fleece was white as snow And everywhere that Mary went Mary went, Mary went \
Everywhere that Mary went The lamb was sure to go" #Se guarda la cadena de texto en la variable 'G'
G.find("snow") #Se busca la palabra "snow" dentro de la cadena y se devuelve la posición de la primera letra de la sub-cadena "snow" que está en 'G'
```

En la variable <code>G</code>, reemplaza la sub-cadena <code>Mary</code> con <code>Bob</code>:

```python
H = G.replace('Mary', 'Bob') #Se reemplaza la sub-cadena "Mary" por "Bob" en la cadena orginal 'G'
H #Se muestra la variable 'H' que contiene el reemplazo hecho en 'G'
```
