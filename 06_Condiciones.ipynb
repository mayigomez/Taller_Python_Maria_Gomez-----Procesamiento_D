<h3>Introducción </h3>


 <div style="text-align: justify;">Las estructuras condicionales permiten ejecutar diferentes bloques de código según el resultado de una condición lógica. En Python, se utilizan sentencias como if, elif y else para controlar el flujo de ejecución de un programa.
En este cuaderno se analiza cómo funcionan las condiciones y cómo se pueden combinar con operadores lógicos y relacionales para tomar decisiones dentro del código.</div>

<h4> Solución del Cuestionario sobre Condiciones </h4>

Escribe una declaración if que determine si un disco musical tiene una calificación mayor a 8. Compruébalo utilizando el calificación del disco <b>“Back in Black”</b> el cual tiene un valor de 8.5. Si la condición es verdadera, imprime “"This album is Amazing!"”

```python
album__rating = 8.5 #Se guarda en la variable 'album__rating' el valor de la calificación del disco Back in Black
if album__rating > 8:  #Condición que evalúa si la calificación del disco es mayor a 8
    print("This album is Amazing!") #Si la condición es verdadera, se imprime un mensaje "This album is Amazing!"
```

Escribe una declaración if-else que determine lo siguiente. Si la calificación es mayor que 8 imprime “this album is amazing”. Si es menor o igual a 8 que imprima “this album is ok”.

```python
if album__rating > 8:  #Condición que evalúa si la calificación del disco es mayor a 8
    print("This album is Amazing!") #Si la condición es verdadera, se imprime un mensaje "This album is Amazing!"
else:
    print("This album is ok.") #De lo contrario, si la condición es falsa, se imprime un mensaje "This album is ok."
```
