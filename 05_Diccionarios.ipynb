<b>Diccionario para las siguientes dos preguntas:</b>

```python
# Pregunta - Diccionario de ejemplo

soundtrack_dic = {"The Bodyguard":"1992", "Saturday Night Fever":"1977"}
soundtrack_dic 
```

a) En el diccionario <code>soundtrack_dict</code>. ¿Cuáles son las llaves?

```python
soundtrack_dic.keys() #Se pretende encontrar cuáles son las llaves que contiene el diccionario 'soundtrack_dic'
```

b) En el diccionario <code>soundtrack_dict</code>. ¿Cuáles son los valores?

```python
soundtrack_dic.values() #Se quiere encontrar cuáles son los valores correspondientes a las llaves del diccionario
```

<b>Diccionario para las siguientes preguntas:</b>


Los discos <b>Back in Black</b>, <b>The Bodyguard</b> y <b>Thriller</b> poseen el siguiente record de ventas en millones, 50, 50 y 65 respectivamente:

a) Crea un diccionario <code>album_sales_dict</code> donde las llaves sean el nombre del disco y los valores la cantidad de ventas en millones.

```python
#Se crea un diccionario con los álbumes y sus ventas correspondientes
album_sales_dict = {"Back in Black": 50, "The Bodyguard": 50, "Thriller": 65} 

#Se visualiza como se guardaron los elementos dados, en el  diccionario yque estos sean correspondientes a lo solicitado
album_sales_dict
```

b) Usa el diccionario para encontrar las ventas totales de <b>Thriller</b>:

```python
album_sales_dict['Thriller'] #Se obtine el valor de la venta total del album a partir de su llave, Thriller
```

c) Encuentra los nombres de los discos dentro del diccionario mediante el método <code>keys</code>:

```python
album_sales_dict.keys() #Se busca mediante las llaves cuáles son los álbumes/discos que se encuentran en el diccionario
```

d) Encuentra los records de ventas dentro del diccionario mediante el método <code>values</code>:

```python
album_sales_dict.values() #Se busca mediante los valores cuáles son las ventas totales
                          #de los álbumes/discos que se encuentran en el diccionario
```
