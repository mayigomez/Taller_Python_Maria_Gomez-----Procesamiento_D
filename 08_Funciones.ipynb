<h3>Introducción </h3>


 <div style="text-align: justify;">Las funciones en Python son bloques de código que permiten agrupar instrucciones para realizar una tarea específica. Pueden recibir parámetros y devolver resultados, lo que facilita la reutilización del código y mejora su organización.
En este cuaderno se estudia la definición de funciones, el uso de parámetros, valores de retorno y el alcance de las variables dentro y fuera de una función.</div>

<h4> Solución del Cuestionario sobre Funciones </h4>

Cree una función que divida la primera entrada por la segunda entrada:

```python
def div(a, b):
    return(a/b)
#Ejemplo de su uso 
#Se define un valor numérico para las variables a y b
a = 10
b = 21
print(div(a, b)) #Se llama la función div para los nuevos valores de a y b y se imprime el resultado
```

Utilice la función <code> con </code> para la siguiente pregunta.

```python
def con(a, b):
    return(a + b)
```

¿Se puede usar la función <code> con </code> que definimos antes para agregar números enteros o cadenas?

```python
# Se intenta con números
a = 4
b = 12
print(con(a, b)) #Se llama la función con para los números dados y se imprime el resultado

#El contenido de las variables a y b se cambia a None para asignarles nuevos valores de tipo cadena
a = None
b = None

# Se intenta con cadenas
a = "Mario"
b = " Bros"
print(con(a, b)) #Se llama a la función con para las cadenas dadas y se imprime el resultado 
```

¿Se puede usar la función <code> con </code> que definimos antes para concentrar una lista o tupla?

```python
a = [1, 2, 3]
b = [1.5, 2.5, 3.5]
print(con(a, b)) #Se llama a la función con para las listas dadas para a y b; y se imprime el resultado
```
