Transformar la lista <code>['rap','house','electronic music', 'rap']</code> en un conjunto:

```python
album_set = set(["rap","house","electronic music", "rap"]) #Se crea un conjunto y se le asigna a la variable
                                                           #album_set los elementos propuestos en el enunciado

album_set #Se muestra el contenido guardado previamente en la variable, sin el duplicado de rap que estaba 
          #presente en el conjunto, ya que los conjuntos no permiten elementos repetidos y se elimina automáticamente.
```

Considera que la lista <code>A = [1, 2, 2, 1]</code> y el conjunto <code>B = set([1, 2, 2, 1])</code> suman <code>sum(A) = sum(B)</code>

```python
Number_A = [1,2,2,1] #Se crea una lista y se le asigna a la variable Numbre_A con los elementos 1,2,2,1
Number_B = set([1,2,2,1]) #Se crea un conjunto y se le asigna a la variable Number_B con los elementos 1,2,2,1

print("Contenido de la lista Number_A:", Number_A) #Se muestra el contenido de la lista Number_A
print("Contenido del conjunto Number_B:", Number_B) #Se muestra el contenido del conjunto Number_B

#Se opera la suma de los elementos de cada varianble independientemente con el 
# fin de verificar que el resultado de la operación sea igual en ambas variables
print("\nSuma de los elementos de la lista Number_A:", sum(Number_A)) 
print("Suma de los elementos del conjunto Number_B:", sum(Number_B))
```

Crea un nuevo conjunto <code>album_set3</code> que sea la unión de <code>album_set1</code> y <code>album_set2</code>:

```python
#Se crean dos conjuntos y se le asignan a las variables album_set1 y album_set2
album_set1 = set(["Thriller", 'AC/DC', 'Back in Black']) 
album_set2 = set([ "AC/DC", "Back in Black", "The Dark Side of the Moon"])

#Se muestra el contenido de los conjuntos album_set1 y album_set2
print("Contenido de album_set1: ",album_set1)
print("Contenido de album_set2: ", album_set2)

#Se quiere hacer la unión de los conjuntos album_set1 y album_set2 y guardar el resultado en album_set3
album_set3 = album_set1.union(album_set2)
print(" \nUnión de los conjuntos album_set1 y album_set2: ",album_set3) #Se muestra el resultado final de la unión de los dos conjuntos
```

Averigua si <code>album_set1</code> es un subconjunto de <code>album_set3</code>:

```python
set(album_set1).issubset(album_set3)     
```
