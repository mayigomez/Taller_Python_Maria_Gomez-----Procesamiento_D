Escribe un bucle <code>for</code> que imprima todos los elementos entre <b>-5</b> y <b>5</b> usando la función range.

```python
for i in range(-5, 6): # Se crea un bucle for que itere los números del -5 al 5
    print(i) #Se imprimen los números que se encuentran en el rango (-5, 5)
```

Imprime los elementos de la siguiente lista:
<code>Genres=[ 'rock', 'R&B', 'Soundtrack', 'R&B', 'soul', 'pop']</code>
Y asegúrate de seguir las convenciones del lenguaje Python

```python
Genres = ["rock", "R&B","Soundtrack", "R&B", "soul", "pop"] #Se crea una lista Genres que guarde géneros musicales y sus duplicados
for Genre in Genres: #Se crea un bucle for que itere los géneros de la lista Genres
    print(Genre) #Se imprimen todos lo géneros musicales iterados de la lista Genres
```

Escribe un bucle for que imprima la siguiente lista: <code>squares=['red', 'yellow', 'green', 'purple', 'blue']</code>

```python
squares = ["red", "yellow", "green", "purple", "blue"] #Se crea una lista squares que guarde los colores dados
for square in squares: #Se crea un bucle for con el fin que itere los elementos (colores) de la lista squares
    print(square) #Se imprimen los colores iterados de la lista squares
```

Escribe un bucle while que muestre los valores de la calificación de las canciones de un disco almacenados en la lista <code>PlayListRatings</code>. Si la calificación es menor a 6, sal del bucle. La lista <code>PlayListRatings</code> esta compuesta por: <code>PlayListRatings = [10, 9.5, 10, 8, 7.5, 5, 10, 10]</code>

```python
PlayListRatings = [10, 9.5, 10, 8, 7.5, 5, 10, 10] #Se crea una lista llamada PlayListRatings que guarde
                                                   #las calificaciones de las canciones de un discos/álbum
i = 0  #Se inicializa la variable i en cero para usarlo como índice que itere la lista PlayListRatings en el bucle while

#Se crea un bucle while que itere la lista PlayListRating mientras i (índice) sea menor que la longitud de la lista
#y la calificación de la canción en la posición i sea mayor o igual a 6; y si no cumple una de las dos se sale del bucle
while (i < len(PlayListRatings) and PlayListRatings[i] >= 6): 
    print(PlayListRatings[i])      #Mientras el bucle esté en ejecución se imprimiran las calificaciones de las canciones en i
    i = i +1          #A i se le incrementa 1 para que el bucle itere el siguiente elemento de la lista
```

Escribe un bucle while que copie las cadenas <code>'orange'</code> de la lista <code>squares</code> a la lista <code>new_squares</code>. Detente y sal del ciclo si el valor en la lista no es <code>'orange'</code>:

```python
squares = ['orange', 'orange', 'purple', 'blue ', 'orange'] #Se crea una lista squares que guarde los colores dados
new_squares = [] #Se crea una lista vacía llamada new_squares para guardar las cadenas que cumplan la condición del bucle while
i = 0 #Se inicializa la variable i en cero para usarlo como índice que itere la lista squares en el bucle while


#Se crea un bucle while que itere la lista square mientras i (índice) sea menor que la longitud de la lista
#  y el color en la posición i sea igual a 'orange'; y si no cumple una de las dos condiciones se termina del ciclo
while (i < len(squares) and squares[i] == 'orange'):
    new_squares.append(squares[i]) #Si el bucle se encuentra en ejecución se agrega a la lista new_squares la cadena en i
    i += 1    #A i se le incrementa 1 para que el bucle itere el siguiente elemento de la lista


print(new_squares) #Se imprime la lista new_squares desbués de que el ciclo while allá terminado su ejecución
```
