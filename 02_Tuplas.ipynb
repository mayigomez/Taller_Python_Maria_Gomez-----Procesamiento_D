<h4> Solución del Examen sobre Tuplas </h4>

Considérese la siguiente tupla:

```python
# Tupla de ejemplo

genres_tuple = ("pop", "rock", "soul", "hard rock", "soft rock", \
                "R&B", "progressive rock", "disco") #Se guarda una tupla con diferentes palabras (géneros de música)
                                                    # en la variable genres_tuple
genres_tuple
```

Encontrar la longitud de la tupla, <code>genres_tuple</code>:

```python
len(genres_tuple) #Se busca saber cual es la longitud de la tupla previamnete guardada en genres_tuple
```

Accede al elemento en el índice 3:

```python
genres_tuple[3] #Se busca el elemento de la tupla ubicado en índice 3
```

Se pudo utilizar el - tbm

Utiliza el slicing para obtener los índices 3, 4 y 5:

```python
genres_tuple[3:6] #Se busca saber cuáles son los géneros musicales que se encuentran entre los índices 3-5
```

Encuentra los primeros dos elementos de la tupla <code>genres_tuple</code>:

```python
genres_tuple[0:2] #Con esto se pretende encontrar cuáles son los dos primeros géneros ubicados en la tupla
```

Encuentra el primer índice de <code>"disco"</code>:

```python
genres_tuple.index("disco") #Se busca el índice del elemento "disco" dentro de la tupla
```

Genera una lista ordenada desde la tupla <code>C_tuple=(-5, 1, -3)</code>:

```python
C_tuple = (-5, 1, -3) #Se crea una nueva tupla 'C_tuple' que gusarde los elementos -5, 1 y -3
C_tuple_sorted = sorted(C_tuple)  #Luego, se quiere ordenar los valores que contiene dicha tupla
C_tuple_sorted #Tupla que contiene los elementos ordenados de C_tuple
```
