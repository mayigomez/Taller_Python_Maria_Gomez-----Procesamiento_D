<h3>Introducción</h3>

 <div style="text-align: justify;"> Las listas son una de las estructuras de datos más utilizadas en Python gracias a su flexibilidad y facilidad en su uso, ya que, permiten almacenar múltiples elementos de manera ordenada y modificable. A diferencia de las tuplas, las listas pueden cambiar su contenido, lo que permite agregar, eliminar o modificar elementos durante la ejecución del programa.
En este cuaderno se trabajarán las operaciones básicas sobre listas, así como los métodos más comunes para su manipulación.</div>

<h4> Solución del Examen sobre Listas </h4>

Crea una lista <code>a_list</code> con los siguientes elementos <code>1</code>, <code>hello</code>, <code>[1,2,3]</code> y <code>True</code>.

```python
a_list = [1, "hello", [1,2,3],True] #Se crea una lista y se le asigna a la variable 'a_list'
                                     #con los elementos que propone el enunciado
                                    
a_list #Se muestra el contenido previamente guardado en la variable 'a_list'
```

Encuentra el valor almacenado en el índice 1 de <code>a_list</code>.

```python
a_list[1] #Se busca y se muestra el elemento almacenado en la posición #1 de la lista 'a_list'
```

Obtén el elemento almacenado en el índice 1, 2 y 3 de <code>a_list</code>.

```python
a_list[1:4] #Se busca y se muestra la sub-lista que contiene los elementos que se encuentran entre la posición #1 y #3 de la lista
```

Concatena las siguientes listas <code>A = [1, 'a']</code> y <code>B = [2, 1, 'd']</code>:

```python
A = [1,"a"] #Se guarda en la variable A la lista que contiene los elementos asignados en el enunciado
B = [2,1,"b"] # Se guarda en la variable B la lista que contiene los elementos asignados en el enunciado
# Se muestra el contenido inicial de las listas A y B
print("Contenido inicial de las listas A y B :\n", "Lista A :", A, "\n Lista B:", B, "\n")

C = A + B # Se crea una lista y se le asigna a C la concatenación de las listas A y B
print("Contenido final de la concatenación de las listas A y B:", C) #Se muestra el contenido final de C
```
